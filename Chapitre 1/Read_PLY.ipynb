```python
import open3d
from pytorch3d.io import load_ply
import plotly.graph_objects as go
```

```python


mesh_file = "cow.ply"
print('visualisation du maillage a laide dopen3D')
mesh = open3d.io.read_triangle_mesh(mesh_file)
open3d.visualization.draw_geometries([mesh], mesh_show_wireframe = True, mesh_show_back_face = True, )
```

```python

print("Chargement du même fichier avec PyTorch3D")
vertices, faces = load_ply(mesh_file)
print('Type de sommets = ', type(vertices), ", type de  faces = ", type(faces))
print('sommets = ', vertices.shape)
print('faces = ', faces.shape)
```

```python
fig = go.Figure(
    data=[
        go.Mesh3d(
            x=vertices[:,0],
            y=vertices[:,1],
            z=vertices[:,2],
            i=faces[:,0],
            j=faces[:,1],
            k=faces[:,2],
            opacity=0.50)
    ],
    layout=dict(
        scene=dict(
            xaxis=dict(visible=False),
            yaxis=dict(visible=False),
            zaxis=dict(visible=False)
        )
    )
)
fig.show()
```
